```python
import numpy as np
np.random.seed(0)

import pandas as pd

def convert_timepoints_to_flux( timepoints_ref ):
    timepoints = timepoints_ref.copy()
    return pd.DataFrame( timepoints.loc[:, timepoints.columns[1:]].values - timepoints.loc[:, timepoints.columns[:-1]].values,
                         index = timepoints.index, columns = timepoints.columns[1:])

%matplotlib inline
```

## Timeseries flux of metabolites for PCM-ACO-MFS

To convert metabolite concentrations in $g/L$ to metabolite fluxes in $g/L/day$ we must weight the difference in concentration by the number of hours between measurements

$$v_{t_n} = \left(c_{t_n} -c_{t_{n-1}}\right)\left(\frac{24}{t_{n} - t_{n-1}}\right)$$
where $n$ ranges from timepoint 0 to timepoint $N$, $t_n$ is the number of hours at timepoint $n$, $c_{t_n}$ is the concentration at timepoint $n$, and $v_{t_n}$ is the flux at timepoint $n$.  If the number of concentration timepoints range from $0..N$, the number of flux timepoints range from $1..N$.

```python
table4_mean_conc = table4.loc[ idx[ :, 'mean' ], : ]                      # Generate a table that only contains mean concentrations.
table4_mean_conc.index = table4_mean_conc.index.droplevel('mean-or-std')  # drop the mean-or-std row index
table4_mean_conc.index.name = 'Metabolite $g/L$'
timepoints = np.array( [ int( col[:-1].strip() )                          # Generate timepoints from column index
                        for col in table4_mean_conc.columns] ) 
table4_flux = convert_timepoints_to_flux( table4_mean_conc.copy() )\
                    .multiply( 24/( timepoints[1:] - timepoints[:-1] ))   # apply the formula converting concentrations and timepoints to fluxes
table4_flux.index.name= 'Metabolite $g/L/day$'                       # Provide informative units for the row index name
table4_flux = table4_flux.fillna(0)                             # convert NaN to 0
table4_flux.to_csv('table4_zhao2019.PCM-ACO-MFS.timeseries.grams_per_liter_per_day.tab', 
                         sep='\t')                                        # write out the dataframe to a file.
table4_flux
```

## Map measured metabolites to model metabolites

```python
measured_metabolites = table4_mean_conc.join(table3_mean_conc).T
measured_metabolites
```

According to Table3 of [Zhao 2019], all metabolite measurements were extracelluar.

According to [Rafelski 2012](https://f1000.com/fulltext/doi/10.1126/science.1225720), the ratio of cytosol to mitochondrion in yeast is 100:1, so total concentration is mostly cytosolic, and estimates of mitochondrial concentrations would be unreliable.

Rafelski, S. M., Viana, M. P., Zhang, Y., Chan, Y.-H. M., Thorn, K. S., Yam, P., Fung, J. C., Li, H., Costa, L. da F. and Marshall, W. F. (2012). Mitochondrial network size scaling in budding yeast. Science 338(6108), 822824.

```python
metabolite_map = {
 'pyr_e': 'Pyruvate',
 'acon_C_e': 'cis-aconitic acid',
 'akg_e': 'alpha-ketoglutaric acid',
 'acon_T_e':    'trans-aconitic acid',
 'cit_e': 'Citric acid',
 'itacon_e': 'Itaconic acid'
}
```

```python
model_metabolites = pd.DataFrame()
for metabolite in metabolite_map:
    model_metabolites[metabolite] = measured_metabolites[metabolite_map[metabolite]]/grams_per_mol[metabolite_map[metabolite]]
model_metabolites.index.name = 'Metabolite $mol/L$'
model_metabolites = model_metabolites.sort_index(axis=1)
model_metabolites.drop('Po1f').drop('20 h').to_csv('glycolysis_itaconate_pwy_metabolites_zhao2019.csv')
model_metabolites
```

### Model reaction to Measured Enzyme map
If multiple enzymes catalyze a reaction, Peter St. John averaged their expression.
We may choose a more sophisticated approach, where we take the min of complex subunits and the sum of isozome expression.
$$
\begin{eqnarray}
  n A  + m B & \rightarrow &  AB \\
Expression(complex)  & = &  \min(A/n, B/m)  \\
 Expression(isozyme) &=& A + B \\
 Expression(multifunctional enzyme) = \sum_{r} 
\end{eqnarray}
$$

Since ACO has a mitochondrial location signal, we assume the measured ACO is predominantly in the mitochondrion.

```python
enzyme_map= {'ACONTa':'ACO',
             'ACONTb':'ACO',
             'MTT':'MTT',
             'ITACON':'CAD',
             'ITACONtr':'MFS',
             'CSm': 'CIT',
             'ICDHym': 'IDH',
             'PPCK': 'PCK',
             'PC': 'PYC'
            }
```

```python
measured_enzymes = enzyme_table4.join(enzyme_table3).T
#measured_enzymes['CAD'] = 1
#measured_enzymes.loc['Po1f','CAD'] = 0
measured_enzymes
```

```python
model_enzymes = pd.DataFrame()
for enzyme in enzyme_map:
    model_enzymes[enzyme] = measured_enzymes[enzyme_map[enzyme]]
model_enzymes.index.name = 'Enzyme fold change'
model_enzymes = model_enzymes.sort_index(axis=1)
model_enzymes = model_enzymes.drop('Po1f').drop('20 h')
model_enzymes.to_csv('glycolysis_itaconate_pwy_enzymes_zhao2019.csv')
model_enzymes
```

## Map measured fluxes to model reactions

```python
measured_flux = table4_flux.join(table3_flux).T
measured_flux['Glucose'] = 21
measured_flux.loc['44 h','Glucose'] = 25
measured_flux.loc[['102 h', '201 h', '308 h', '386 h', '460 h'],'Glucose'] = 0
measured_flux
```

```python
def model_flux_from_measured_flux( measured_flux, flux_map, grams_per_mol, units =  'Metabolite flux $mol/L/day$' ):
    model_flux = pd.DataFrame()
    for rxn in flux_map:
        model_flux[rxn] = measured_flux[flux_map[rxn]]/grams_per_mol[flux_map[rxn]]
    model_flux = model_flux.sort_index(axis=1)
    model_flux.index.name = units
    return model_flux
flux_map= {
 'EX_itacon_e': 'Itaconic acid',
 'EX_cit_e':    'Citric acid',
 'EX_akg_e':   'alpha-ketoglutaric acid',
 'EX_acon_C_e': 'cis-aconitic acid',
 'EX_glc_D_e':  'Glucose' , 
 'EX_pyr_e':    'Pyruvate',
 'EX_acon_T_e': 'trans-aconitic acid'
}

grams_per_mol = {'Itaconic acid':128.08,
                 'Citric acid': 189.101,
                 'alpha-ketoglutaric acid':144.084,
                 'cis-aconitic acid': 171.09,
                 'Glucose': 180.157,
                 'Pyruvate': 87.055,
                 'trans-aconitic acid': 171.09,
}

model_flux = model_flux_from_measured_flux( measured_flux, flux_map, grams_per_mol )
                    

model_flux.drop('Po1f').to_csv('glycolysis_itaconate_pwy_flux_zhao2019.csv')
model_flux
```
