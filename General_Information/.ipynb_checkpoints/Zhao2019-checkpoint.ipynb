# Table 3 from Zhao 2019

```python
import pandas as pd
table3 = pd.read_html('table3.txt',index_col='Strain',na_values='ND')[0]
columns = table3.columns 
for col in table3.columns:
    table3['{}-mean'.format(col)] = table3[col].str.split(r'[^\d\.]+').str.get(0)
    table3['{}-std'.format(col)] = table3[col].str.split(r'[^\d\.]+').str.get(-1)
table3 = table3.drop(columns, axis=1 )
for protein in ['PCM','PYC','MFS','PCK','MTT','CIT', 'ACO']:
    table3[protein] = [1 if protein in strain else 0.5 for strain in table3.index]
table3
```

```python
t3t = table3.T
t3t['mean-or-std'] = ['std'  if 'std' in idx else 'mean' for idx in t3t.index]
t3t['metabolite'] = ['-'.join(idx.split('-')[0:-1]) for idx in t3t.index]
t3t
```

```python
t3t.to_csv('Table3-Zhao2019.tab', sep='\t')
```
